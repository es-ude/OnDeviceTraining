```python
import torch
from torch import nn, Tensor
from torch.nn import Parameter, MSELoss

# define linear layer
linear = nn.Linear(3, 2)

# assign weights and bias correctly
linear.weight = Parameter(torch.tensor([[5.,  -1., 9.],
                                        [ 22.,  -100., 18.]]))
linear.weight = Parameter(torch.tensor([[1., 1., 1.],
                                        [1., 1., 1.]]))
linear.bias = Parameter(torch.tensor([-1., 3.]))

relu = nn.ReLU()

model = nn.Sequential(linear, relu)

# input must require gradients
x1 = Parameter(torch.tensor([-4., 1., 9.]), requires_grad=True)

x2 = Parameter(torch.tensor([5., -1., 2]), requires_grad=True)

x3 = Parameter(torch.tensor([-7., -5., 6]), requires_grad=True)

label_1 = torch.tensor([ 59., -23.])
label_2 = torch.tensor([ 43., 249.])
label_3 = torch.tensor([ 23., 457.])

optim = torch.optim.SGD(model.parameters(), lr=0.01)
loss_func = nn.MSELoss()
for i in range(1000):
    
    # forward
    y1 = model(x1)
    #print("y =", y1)
    loss = loss_func(y1, label_1)
    loss.backward()
    print(f"1: {loss.item()=}")
    # forward
    y2 = model(x2)
    loss = loss_func(y2, label_2)
    loss.backward()
    print(f"2: {loss.item()=}")
    #print("y =", y2)
    # forward
    y3 = model(x3)
    loss = loss_func(y3, label_3)
    loss.backward()
    print(f"3: {loss.item()=}")
    #print("y =", y3)
    optim.step()
    optim.zero_grad()
    print()
    
    #print({x for x in model.parameters()})
# loss
#loss_fn = MSELoss()
#target = torch.tensor([-4., -3.])

#loss1 = loss_fn(y1, target)

# backward pass #1
#loss.backward(retain_graph=True)
```
